# Definicion del Modelo `Proceso Gausiano`

Definición formal:

Un proceso gaussiano es una colección de variables aleatorias, cualquier número finito de las cuales tiene una distribución gaussiana conjunta. Formalmente, sea X un espacio de índices arbitrario. Un proceso gaussiano en $X$ es una función aleatoria $f: X → ℝ$ tal que para cualquier conjunto finito de puntos ${x₁, ..., xₙ} ⊂ X$, el vector aleatorio $[f(x₁), ..., f(xₙ)]$ tiene una distribución gaussiana multivariada.

Un proceso gaussiano está completamente especificado por su función de media $m(x)$ y su función de covarianza $k(x, x')$:

$$m(x) = E[f(x)]$$

$$k(x, x') = E[(f(x) - m(x))(f(x') - m(x'))]$$

donde $E$ denota la esperanza matemática.

Denotamos un proceso gaussiano como:

$$f(x) ~ GP(m(x), k(x, x'))$$

Explicación detallada:

1. Colección de variables aleatorias: Un GP es una generalización de la distribución gaussiana multivariada a un número infinito de variables. Cada punto en el espacio de entrada $X$ está asociado con una variable aleatoria.

2. Distribución gaussiana conjunta: Cualquier subconjunto finito de estas variables aleatorias sigue una distribución gaussiana multivariada. Esto significa que podemos caracterizar completamente el proceso utilizando solo los dos primeros momentos: la media y la covarianza.

3. Función de media $m(x)$: Esta función representa el valor esperado del proceso en cualquier punto $x$. En muchas aplicaciones, se asume que $m(x) = 0$ para simplificar.

4. Función de covarianza $k(x, x')$: También llamada kernel, esta función define la covarianza entre los valores de la función en diferentes puntos del espacio de entrada. Determina las propiedades de suavidad y variabilidad del proceso.

5. Especificación completa: Un GP queda completamente definido por sus funciones de media y covarianza. Esto significa que para hacer inferencias, solo necesitamos especificar estas dos funciones.

6. Flexibilidad: Los GPs son muy flexibles y pueden modelar una amplia variedad de funciones dependiendo de la elección de la función de covarianza.

7. Inferencia bayesiana: Los GPs proporcionan un marco natural para la inferencia bayesiana. Dados algunos datos observados, podemos calcular la distribución posterior sobre las funciones.

8. Incertidumbre: Una característica clave de los GPs es que proporcionan no solo predicciones puntuales, sino también estimaciones de incertidumbre.

En la práctica, los GPs se utilizan para tareas como regresión, clasificación, optimización y diseño experimental. Son particularmente útiles en situaciones donde se desea modelar funciones suaves con incertidumbre.

```python
from datasetsforecast.losses import mae, mape, mase, rmse, smape

def evaluate_cross_validation(df, metric):
    models = df.drop(columns=['ds', 'cutoff', 'y']).columns.tolist()
    evals = []
    for model in models:
        eval_ = df.groupby(['unique_id', 'cutoff']).apply(lambda x: metric(x['y'].values, x[model].values)).to_frame() # Calculate loss for every unique_id, model and cutoff.
        eval_.columns = [model]
        evals.append(eval_)
    evals = pd.concat(evals, axis=1)
    evals = evals.groupby(['unique_id']).mean(numeric_only=True) # Averages the error metrics for all cutoffs for every combination of model and unique_id
    evals['best_model'] = evals.idxmin(axis=1)
    return evals
```

```python
import pandas as pd
import numpy as np
from scipy import stats



# 1. Método de la demanda promedio
demanda_promedio = opti['Quantity'].mean()
lead_time = 2  # Asumimos un tiempo de entrega de 2 días, ajusta según tus necesidades
inventario_optimo_promedio = demanda_promedio * lead_time

print(f"1. Inventario óptimo (demanda promedio): {inventario_optimo_promedio:.2f}")

# 2. Método de la demanda máxima
demanda_maxima = opti['Quantity'].max()
inventario_optimo_maximo = demanda_maxima * lead_time

print(f"2. Inventario óptimo (demanda máxima): {inventario_optimo_maximo:.2f}")

# 3. Método del punto de reorden (ROP) con stock de seguridad
demanda_diaria = opti['Quantity'].mean()
desviacion_estandar = opti['Quantity'].std()
nivel_servicio = 0.95  # 95% nivel de servicio
z_score = stats.norm.ppf(nivel_servicio)

stock_seguridad = z_score * desviacion_estandar * np.sqrt(lead_time)
ROP = (demanda_diaria * lead_time) + stock_seguridad

print(f"3. Punto de Reorden (ROP): {ROP:.2f}")
print(f"   Stock de Seguridad: {stock_seguridad:.2f}")

# 4. Método de la cantidad económica de pedido (EOQ)
demanda_anual = opti['Quantity'].sum() * (365 / len(opti))  # Estimación de la demanda anual
costo_pedido = 100  # Costo de hacer un pedido, ajusta según tus necesidades
costo_mantenimiento = 0.2  # 20% del costo del producto como costo de mantenimiento anual

precio_promedio = opti['Price per Unit'].mean()
costo_mantenimiento_unidad = precio_promedio * costo_mantenimiento

EOQ = np.sqrt((2 * demanda_anual * costo_pedido) / costo_mantenimiento_unidad)

print(f"4. Cantidad Económica de Pedido (EOQ): {EOQ:.2f}")

# Análisis adicional: Frecuencia óptima de pedidos
frecuencia_optima = demanda_anual / EOQ
print(f"   Frecuencia óptima de pedidos: {frecuencia_optima:.2f} veces al año")

# Visualización de la demanda y los niveles de inventario
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(opti['ds'], opti['Quantity'], label='Demanda real')
plt.axhline(y=inventario_optimo_promedio, color='r', linestyle='--', label='Inventario óptimo (promedio)')
plt.axhline(y=ROP, color='g', linestyle='--', label='Punto de Reorden (ROP)')
plt.axhline(y=EOQ, color='y', linestyle='--', label='EOQ')
plt.title('Demanda vs. Niveles de Inventario Estimados')
plt.xlabel('Fecha')
plt.ylabel('Cantidad')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Análisis de costos
costo_total_EOQ = (demanda_anual / EOQ) * costo_pedido + (EOQ / 2) * costo_mantenimiento_unidad
print(f"\nCosto total anual estimado (usando EOQ): ${costo_total_EOQ:.2f}")

# Recomendaciones
print("\nRecomendaciones para la gestión de inventario:")
print(f"1. Mantener un inventario mínimo de {ROP:.2f} unidades.")
print(f"2. Realizar pedidos de {EOQ:.2f} unidades aproximadamente {frecuencia_optima:.2f} veces al año.")
print(f"3. Mantener un stock de seguridad de {stock_seguridad:.2f} unidades para cubrir fluctuaciones en la demanda.")
print("4. Monitorear constantemente la demanda real y ajustar estos parámetros según sea necesario.")
```
