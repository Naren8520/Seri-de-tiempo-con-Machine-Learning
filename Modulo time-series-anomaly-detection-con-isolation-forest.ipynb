# **Isolation Forest para la Detección de Anomalías: Revelando Perspectivas Ocultas**

En el ámbito de la detección de anomalías, el algoritmo Isolation Forest se destaca como una herramienta poderosa y eficiente para identificar valores atípicos y anomalías dentro de conjuntos de datos. Desarrollado por Liu et al. en 2008, Isolation Forest aprovecha los principios de los árboles de decisión para aislar anomalías de manera efectiva, lo que lo hace especialmente adecuado para detectar anomalías en conjuntos de datos grandes y complejos. Adentremos en cómo funciona Isolation Forest y por qué se ha convertido en una elección popular para tareas de detección de anomalías.

#### **Comprendiendo Isolation Forest:**

1. **Enfoque Basado en Árboles:** Isolation Forest adopta un enfoque diferente en comparación con métodos tradicionales como clustering o técnicas basadas en distancias. Construye un conjunto de árboles de aislamiento para aislar anomalías mediante particionado recursivo.

2. **Principio de Aislamiento:** La idea clave detrás de Isolation Forest es que las anomalías son más fáciles de aislar porque son escasas y tienen valores de atributo que difieren significativamente de las instancias normales. Al identificar estas anomalías primero, Isolation Forest puede detectarlas de manera eficiente.

3. **Longitud del Camino:** Las anomalías se aíslan más rápidamente en un Isolation Forest. La longitud promedio del camino para aislar una anomalía es más corta en comparación con una instancia normal, lo que lo convierte en un método robusto y eficiente para la detección de anomalías.


#### **Aplicaciones de Isolation Forest:**

1. **Ciberseguridad:** Isolation Forest se utiliza ampliamente en ciberseguridad para identificar patrones de tráfico de red inusuales, que podrían indicar ciberataques o brechas de seguridad.

2. **Detección de Fraudes:** En el sector financiero, Isolation Forest puede ayudar a detectar actividades fraudulentas al identificar transacciones o comportamientos inusuales que se desvían de los patrones normales.

3. **Mantenimiento Predictivo:** Isolation Forest se puede utilizar en mantenimiento predictivo para identificar anomalías en datos de sensores de maquinaria o equipos, ayudando a prevenir fallas inesperadas.

#### **Desafíos y Consideraciones:**

1. **Ajuste de Parámetros:** Seleccionar los parámetros correctos, como el número de árboles y el tamaño de la submuestra, es crucial para el rendimiento de Isolation Forest.

2. **Interpretabilidad:** Si bien Isolation Forest es efectivo para detectar anomalías, explicar por qué una instancia en particular se marca como valor atípico podría ser desafiante debido a su naturaleza de caja negra.

En conclusión, Isolation Forest ha surgido como un activo valioso en el campo de la detección de anomalías, ofreciendo una solución robusta y escalable para identificar valores atípicos en conjuntos de datos diversos. Al aprovechar el principio de aislamiento y la estructura basada en árboles, Isolation Forest proporciona un enfoque único para la detección de anomalías que es especialmente adecuado para los desafíos modernos de análisis de datos.

Isolation Forest es una técnica de detección de anomalías no supervisada que se destaca en la identificación de valores atípicos dentro de los datos, lo que la hace adecuada para el análisis de series temporales. Funciona bajo el supuesto de que las anomalías (puntos de datos inusuales) tienden a ser raras y distintas de la mayoría de los datos. Esto permite que el algoritmo las aísle de manera efectiva.

<blockquote style="margin-right:auto; margin-left:auto; background-color: #008080; padding: 1em; margin:3px;font-size:20px;font-family:Comic Sans Ms;color:#660000;letter-spacing:0.7px;text-align: justify;">

**Idea clave: aislamiento por división**

1. El bosque de aislamiento crea una colección de árboles de decisión (un conjunto).
2. Durante la construcción del árbol, cada punto de datos se divide iterativamente en función de características elegidas al azar y valores divididos.
3. Los puntos de datos que se desvían significativamente de la norma (anomalías) son más fáciles de aislar, ya que requieren menos divisiones para llegar a un nodo de hoja en el árbol. Por el contrario, los puntos de datos normales, que tienden a agruparse, requieren más divisiones para aislarse.

<blockquote style="margin-right:auto; margin-left:auto; background-color: #008080; padding: 1em; margin:3px;font-size:20px;font-family:Comic Sans Ms;color:#660000;letter-spacing:0.7px;text-align: justify;">


**Cálculo de la puntuación de anomalía**

1. Cada punto de datos se evalúa a través de todos los árboles de aislamiento del bosque.
2. Para cada árbol, se registra la longitud de la ruta (número de divisiones necesarias para aislar el punto).
3. Luego, se calcula una puntuación de anomalía para cada punto de datos promediando las longitudes de las rutas de todos los árboles del bosque.
4. Las puntuaciones de anomalía más bajas indican una mayor probabilidad de ser un punto de datos normal, mientras que las puntuaciones más altas sugieren una mayor probabilidad de ser una anomalía.

<blockquote style="margin-right:auto; margin-left:auto; background-color: #008080; padding: 1em; margin:3px;font-size:20px;font-family:Comic Sans Ms;color:#660000;letter-spacing:0.7px;text-align: justify;">

**Aplicación del bosque de aislamiento a los datos de series temporales**

1. Preprocesamiento de datos:

- Asegúrese de que los datos de sus series temporales estén limpios y libres de errores o inconsistencias.
- Maneje los valores faltantes de manera adecuada (por ejemplo, imputación o eliminación).
- Considere escalar los datos si las características tienen rangos significativamente diferentes.

2. Enfoque de ventana deslizante:

- Divida la serie temporal en ventanas de un tamaño fijo (que representen un período de tiempo específico).
- Aplique el bosque de aislamiento a cada ventana de forma independiente.
- Esto permite capturar anomalías localizadas dentro de la serie temporal.
- Experimente con diferentes tamaños de ventana para encontrar el equilibrio óptimo entre sensibilidad y eficiencia computacional.

3. Identificación de anomalías:

- Según los puntajes de anomalía calculados para cada punto de datos dentro de una ventana, puede definir un umbral para determinar qué puntos se consideran anomalías.
- Este umbral se puede establecer en función del conocimiento del dominio, el análisis de datos históricos o los métodos estadísticos.
- Los puntos de datos que exceden el umbral se marcan como anomalías potenciales para mayor investigación.

<blockquote style="margin-right:auto; margin-left:auto; background-color: #008080; padding: 1em; margin:3px;font-size:20px;font-family:Comic Sans Ms;color:#660000;letter-spacing:0.7px;text-align: justify;">

**Ventajas de Isolation Forest para la detección de anomalías en series temporales**

1. Aprendizaje no supervisado: no se requieren datos etiquetados para anomalías, lo que lo hace ideal para situaciones en las que los datos etiquetados pueden ser escasos o costosos de obtener.
2. Rápido y eficiente: el algoritmo opera en una complejidad de tiempo lineal, lo que le permite manejar grandes conjuntos de datos de series temporales de manera eficaz.
3. Robusto para alta dimensionalidad: Isolation Forest puede funcionar bien con datos de series temporales de alta dimensión, donde intervienen múltiples características (por ejemplo, lecturas de sensores).
4. Sin medidas de distancia o densidad: el algoritmo no se basa en métricas de distancia predefinidas ni en suposiciones sobre la distribución de datos.

<blockquote style="margin-right:auto; margin-left:auto; background-color: #008080; padding: 1em; margin:3px;font-size:20px;font-family:Comic Sans Ms;color:#660000;letter-spacing:0.7px;text-align: justify;">

**Posibles limitaciones**

1. Selección del umbral: la elección del umbral óptimo para la detección de anomalías puede ser subjetiva y puede requerir experiencia en el dominio o experimentación.
2. Sensibilidad al tamaño de la ventana: el tamaño de la ventana deslizante puede afectar la detección de anomalías. Las ventanas más grandes pueden pasar por alto anomalías de corta duración, mientras que las ventanas muy pequeñas pueden ser sensibles al ruido.

Además, visualizaremos los resultados respectivos y la estructura de los árboles de ejemplo correspondientes.


**El algoritmo de bosque de aislamiento:** De la [documentación de scikit-learn](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.IsolationForest.html#sklearn.ensemble.IsolationForest.decision_function): "El IsolationForest 'aísla' las observaciones seleccionando aleatoriamente una característica y luego seleccionando aleatoriamente un valor dividido entre los valores máximo y mínimo de la característica seleccionada. Dado que la partición recursiva se puede representar mediante una estructura de árbol, la cantidad de divisiones necesarias para aislar una muestra es equivalente a la longitud de la ruta desde el nodo raíz hasta el nodo de terminación. Esta longitud de ruta, promediada sobre un bosque de tales árboles aleatorios, es una medida de normalidad y nuestra función de decisión. La partición aleatoria produce rutas notablemente más cortas para las anomalías. Por lo tanto, cuando un bosque de árboles aleatorios produce colectivamente longitudes de ruta más cortas para muestras particulares, es muy probable que sean anomalías".

# **Modelado**
Primero dividimos nuestros datos en conjuntos de entrenamiento y prueba y luego solo utilizaremos el conjunto de entrenamiento para el entrenamiento del modelo y las decisiones de diseño algorítmico:

Aca pueden haber varias opciones:

3. Realizar un resample por horas

La selección dependera del problema de negocio que se desee observar o conocer 

```python
df_model1 = df.set_index('timestamp').resample('D').mean().reset_index()
df_model1
```

```python
# remuestrear datos y dividirlos en conjuntos de entrenamiento (~2/3 de los datos) y de prueba (~1/3 de los datos)

df_train, df_test = np.split(df_model1, [int(0.65 *len(df_model1))])
```

```python
# Histogram of 'price'
plt.figure(figsize=(18, 6))
sns.histplot(df_train["value"], bins=20, kde=True, color='blue')
plt.title('Value Distribution Training Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

Teniendo en mente las 3 anomalías de nuestro conjunto de entrenamiento, ya podemos determinar que las anomalías que queremos detectar se encuentran en las colas de la distribución de valores. Ahora podemos utilizar este hallazgo para decidir que debería ser suficiente utilizar los valores inalterados como característica en nuestro modelo y que no debería ser necesaria la ingeniería de características real para tener éxito en nuestra tarea.

Estimamos que la contaminación (la proporción de valores atípicos en el conjunto de datos) de nuestros datos es 3/139 (#anomalías en el conjunto de entrenamiento/#puntos de datos en el conjunto de entrenamiento) respectivamente después de redondear aproximadamente 0,03. Con esto, entrenamos el modelo y obtenemos las predicciones y los puntajes de anomalía de nuestro modelo. Los puntajes de anomalía se utilizarán más adelante para ajustar el modelo.

```python
features = ['value']
df_train1 = df_train[features]
df_test1 = df_test[features]
```

```python
model = IsolationForest(random_state=0, contamination = 0.03)

# Entrenamos el modelo
model.fit(df_train1)
model
```

```python
# Forecasting para determinar las anoalias
outliers_train = pd.Series(model.predict(df_train1)).apply(lambda x: 1 if x == -1 else 0).to_numpy()
outliers_train
```

- ## Decision Function
Puntuación de anomalía promedio de X de los clasificadores base.

La puntuación de anomalía de una muestra de entrada se calcula como la puntuación de anomalía promedio de los árboles del bosque.

La medida de normalidad de una observación dado un árbol es la profundidad de la hoja que contiene esta observación, que es equivalente a la cantidad de divisiones necesarias para aislar este punto. En caso de que queden varias observaciones n_left en la hoja, se suma la longitud de ruta promedio de un árbol de aislamiento de n_left muestras.

```python
# Puntuación de las anomalias 
anomaly_score_train = model.decision_function(df_train1)
anomaly_score_train
```

```python
# agregar resultados de predicción al marco de datos para fines de visualización
df_train_split1 = df_train.assign(outliers = outliers_train)
df_train_split1 = df_train_split1.assign(anomaly_score = anomaly_score_train)
df_train_split1.head()
```

Para reducir la cantidad de anomalías activadas (que podrían ser interesantes, por ejemplo, cuando solo desea una detección algorítmica por evento anormal), ahora ajustaremos el límite de decisión del modelo utilizando un umbral en los puntajes de anomalía del modelo. Esto evitará el reentrenamiento del modelo para proteger los recursos y poder iterar más rápido.

Los puntajes de anomalía del modelo de bosque de aislamiento son devueltos por el método `decision_function`. La documentación de scikit-learn dice [lo siguiente](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.IsolationForest.html#sklearn.ensemble.IsolationForest.decision_function) sobre los puntajes: "Cuanto más bajos, más anormales. Los puntajes negativos representan valores atípicos (anomalías), los puntajes positivos representan valores atípicos".

Por lo tanto, el umbral predeterminado que utiliza el modelo scikit-learn es 0. Ahora corregiremos este umbral hacia abajo para eliminar dos de las tres anomalías activadas en julio.

Primero, tracemos la distribución de los puntajes de anomalía para obtener una descripción general:

```python
frequencies, edges = np.histogram(anomaly_score_train, 50)
hv.Histogram((edges, frequencies)).opts(width=800, height=300,tools=['hover'], xlabel='Anomaly Score')
```

Para determinar con precisión el umbral, tracemos una lista de todas las puntuaciones negativas:

```python
df_train_split1[df_train_split1.anomaly_score < 0]
```

Nos gustaría eliminar las anomalías del 4 y 5 de julio de 2014. Para ello, podemos, por ejemplo, fijar el umbral en -0,017:

```python
threshold = -0.017
hover = HoverTool(tooltips=tooltips)
hv.Points(df_train_split1.query("anomaly_score < {}".format(threshold))).opts(size=8, 
                                                                              color='#ff0000') * hv.Curve(df_train_split1).opts(opts.Curve(title = "Anomalías en el número de pasajeros de los taxis de la ciudad de Nueva York", 
                                                                                                                                                   xlabel="", ylabel="Number Passengers" , height=300, responsive=True,tools=[hover,'box_select', 'lasso_select', 'tap'],show_grid=True))
```

Eso se ve bien :). Ahora probemos el modelo junto con el umbral ajustado en los datos de prueba para ver qué tan bien se generaliza a los nuevos datos:

```python
outliers_test = pd.Series(model.predict(df_test1)).apply(lambda x: 1 if x == -1 else 0).to_numpy()
anomaly_score_test = model.decision_function(df_test1)

# add prediction results to data for visualization purposes
df_test_split1 = df_test.assign(outliers = outliers_test)
df_test_split1 = df_test_split1.assign(anomaly_score = anomaly_score_test)
df_test_split1.head()
```

```python
df_test_split1[df_test_split1["outliers"] == 1]
```
