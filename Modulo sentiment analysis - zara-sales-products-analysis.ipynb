# Introducción 

**Desatando Perspicacias: La Sinergia del Análisis de Sentimientos con Series Temporales**

En el siempre cambiante panorama del análisis de datos, la fusión del análisis de sentimientos y los datos de series temporales ha surgido como una herramienta poderosa para extraer ideas valiosas y comprender tendencias en diversos ámbitos. Esta sinergia no solo enriquece el análisis de series temporales tradicional, sino que también agrega un toque humano al incorporar sentimientos y emociones en el análisis.

**Entendiendo el Análisis de Sentimientos y los Datos de Series Temporales**

El análisis de sentimientos implica el uso de técnicas de procesamiento de lenguaje natural para determinar el sentimiento o tono emocional detrás de un texto. Clasifica el sentimiento como positivo, negativo o neutral, brindando una comprensión más profunda de las opiniones y actitudes humanas. Por otro lado, los datos de series temporales consisten en observaciones registradas a intervalos regulares a lo largo del tiempo, lo que permite a los analistas identificar patrones, tendencias y variaciones estacionales.

**El Poder de Combinar el Análisis de Sentimientos con los Datos de Series Temporales**

Al integrar el análisis de sentimientos con los datos de series temporales, los analistas pueden desbloquear una gran cantidad de posibilidades:

1. **Mayor Precisión en las Predicciones**: El análisis de sentimientos puede actuar como una característica adicional en los modelos de predicción, proporcionando información contextual que puede impactar las tendencias futuras. Por ejemplo, analizar el sentimiento en las redes sociales hacia una marca junto con los datos de ventas puede mejorar las previsiones de ventas.

2. **Identificación de Impactos de Eventos**: El análisis de sentimientos puede ayudar a identificar eventos significativos o noticias que influyen en las tendencias de los datos de series temporales. Detectar picos en el sentimiento en torno a eventos específicos puede explicar anomalías en los datos de series temporales.

3. **Gestión de Riesgos e Ideas de Mercado**: Monitorear el sentimiento en noticias financieras y en redes sociales puede ofrecer ideas valiosas para la gestión de riesgos y decisiones de inversión. Los cambios en el sentimiento pueden preceder a movimientos en el mercado, ayudando en la toma de decisiones oportuna.

4. **Análisis del Sentimiento del Cliente**: Analizar el sentimiento de los clientes a lo largo del tiempo puede ayudar a las empresas a evaluar la satisfacción del cliente, identificar tendencias en la retroalimentación y adaptar productos o servicios en consecuencia.

**Desafíos y Consideraciones**

Si bien la integración del análisis de sentimientos con los datos de series temporales es poderosa, presenta su propio conjunto de desafíos:

1. **Calidad de los Datos y Ruido**: Asegurar la calidad y relevancia de los datos de sentimiento es crucial, ya que el ruido o las inexactitudes pueden sesgar los resultados del análisis.

2. **Sincronización de Datos**: Alinear los datos de sentimiento con los puntos correspondientes de los datos de series temporales puede ser un desafío, que requiere una cuidadosa preprocesamiento y sincronización de datos.

3. **Interpretación de la Subjetividad**: El análisis de sentimientos implica interpretar emociones humanas subjetivas, lo que puede introducir complejidades en el análisis e interpretación.


La fusión del análisis de sentimientos con los datos de series temporales abre nuevas oportunidades para la toma de decisiones basada en datos en diversos sectores. Al aprovechar el poder tanto de la inteligencia emocional como de las tendencias históricas, las organizaciones pueden obtener ideas más profundas, mejorar la precisión de las predicciones y tomar decisiones informadas. A medida que la tecnología continúa avanzando, aprovechar esta sinergia será clave para mantenerse a la vanguardia en la era de las ideas impulsadas por datos.

# Modelo: Idea general

Para desarrollar un modelo matemático de procesamiento de lenguaje natural (NLP) para determinar el sentimiento o tono emocional detrás de un texto, podemos utilizar técnicas como el análisis de sentimientos. A continuación, presentaré una descripción general del enfoque utilizando una representación simplificada basada en la clasificación binaria de sentimientos (positivo o negativo).

**Representación Matemática del Modelo:**

1. **Preprocesamiento del Texto:**
   - Sea $( D )$ el conjunto de documentos de texto.
   - Cada documento $( d_i )$ en $( D )$ se representa como una secuencia de palabras: $( d_i = (w_{i1}, w_{i2}, ..., w_{in}) )$, donde $( n )$ es el número de palabras en el documento.

2. **Representación del Texto:**
   - Se utiliza una matriz de términos de documentos para representar los documentos.
   - Sea $( X )$ una matriz de términos de documentos donde $( x_{ij} )$ representa la frecuencia de la palabra $( j )$ en el documento $( i )$.
   - También se puede utilizar TF-IDF u otras representaciones vectoriales de palabras.

3. **Modelo de Aprendizaje Automático:**
   - Sea $( y )$ el vector de etiquetas de sentimiento donde $( y_i )$ indica el sentimiento (positivo o negativo) asociado con el documento $( i )$.
   - Se utiliza un clasificador $( f )$ para predecir el sentimiento a partir de la representación del texto: $( f(X) = \hat{y} )$.

4. **Función Objetivo:**
   - Se puede utilizar una función de pérdida como la entropía cruzada para entrenar el modelo de clasificación binaria.

5. **Entrenamiento del Modelo:**
   - Se busca minimizar la función de pérdida a través del proceso de optimización: $( \theta^* = \arg \min_\theta \mathcal{L}(f(X; \theta), y) )$, donde $( \theta )$ son los parámetros del modelo.

6. **Predicción del Sentimiento:**
   - Una vez entrenado el modelo, se puede predecir el sentimiento de un nuevo documento $( d_{new} )$ utilizando el modelo entrenado: $( \hat{y}_{new} = f(X_{new}) )$.

En términos matemáticos, el proceso de entrenamiento implica ajustar los parámetros $( \theta )$ del modelo para minimizar la función de pérdida. Este proceso se realiza mediante técnicas de optimización como el descenso de gradiente.

En resumen, el modelo matemático para determinar el sentimiento o tono emocional detrás de un texto implica representar el texto como datos numéricos, entrenar un clasificador de aprendizaje automático para predecir el sentimiento, y optimizar los parámetros del modelo para minimizar la función de pérdida. Este enfoque es fundamental en el análisis de sentimientos en el procesamiento de lenguaje natural.

# Analisis de Sentimiento

Vamos a usar la función `SentimentIntensityAnalyzer()` la cual la vamos a utilizar para analizar el sentimiento de un texto y asignarle una puntuación de sentimiento que indica la positividad, negatividad, neutralidad y la intensidad del sentimiento. 

En este caso, `SentimentIntensityAnalyzer` es la clase que proporciona la funcionalidad para analizar el sentimiento del texto. La función `polarity_scores()` devuelve un diccionario con las puntuaciones de sentimiento que incluyen la negatividad, neutralidad, positividad y una puntuación compuesta que resume el sentimiento general del texto.

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.util import *

from nltk import tokenize

#nltk.download('vader_lexicon')
```

- ## Variable Nombre(name)

```python
store_zara[["name"]].head(10)
```

```python
sid = SentimentIntensityAnalyzer()

store_zara['sentiment_compound_name'] = store_zara.Name.apply(lambda x:sid.polarity_scores(x)['compound'])
store_zara['sentiment_neutral'] = store_zara.Name.apply(lambda x:sid.polarity_scores(x)['neu'])
store_zara['sentiment_OUTDATED'] = store_zara.Name.apply(lambda x:sid.polarity_scores(x)['neg'])
store_zara['sentiment_TRENDY'] = store_zara.Name.apply(lambda x:sid.polarity_scores(x)['pos'])
store_zara['sentiment_name']=''
store_zara.loc[store_zara.sentiment_compound_name>0,'sentiment_name']='TRENDY'
store_zara.loc[store_zara.sentiment_compound_name==0,'sentiment_name']='OUTDATED' # anticuada
store_zara.loc[store_zara.sentiment_compound_name<0,'sentiment_name']='NEUTRAL'
store_zara.head()
```

En este caso le hemos pedido que analice el text `name` y que agregue un algun tipo de analisis de categorias:

1. Trendy
2. OutDated
3. Neutral

```python
#matplotlib.style.use('ggplot')

zara_sentiment = store_zara.groupby(['sentiment_name'])['sentiment_neutral'].count()
zara_sentiment.rename("",inplace=True)
explode = (0, 0, 1.0)
plt.subplot(221)
zara_sentiment.transpose().plot(kind='barh',figsize=(18, 6))
plt.title('Sentiment Analysis 1', bbox={'facecolor':'0.8', 'pad':0})
plt.subplot(222)
zara_sentiment.plot(kind='pie',figsize=(18, 6),autopct='%1.1f%%',shadow=True,explode=explode)
plt.legend(bbox_to_anchor=(1, 1), loc=3, borderaxespad=0.)
plt.title('Sentiment Analysis 2', bbox={'facecolor':'0.8', 'pad':0})
plt.show()
```

## - Variable de Descripcion

```python
store_zara[["description"]].head(10)
```

```python
sid = SentimentIntensityAnalyzer()

store_zara['sentiment_compound_description'] = store_zara.description.apply(lambda x:sid.polarity_scores(x)['compound'])
store_zara['sentiment_neu'] = store_zara.description.apply(lambda x:sid.polarity_scores(x)['neu'])
store_zara['sentiment_negative'] = store_zara.description.apply(lambda x:sid.polarity_scores(x)['neg'])
store_zara['sentiment_positive'] = store_zara.description.apply(lambda x:sid.polarity_scores(x)['pos'])
store_zara['sentiment_description']=''
store_zara.loc[store_zara.sentiment_compound_description>0,'sentiment_description'] = 'POSITIVE'
store_zara.loc[store_zara.sentiment_compound_description==0,'sentiment_description'] = 'NEUTRAL'
store_zara.loc[store_zara.sentiment_compound_description<0,'sentiment_description'] = 'NEGATIVE'
store_zara.head()
```

```python
zara_sentiment = store_zara.groupby(['sentiment_description'])['sentiment_neutral'].count()
zara_sentiment.rename("",inplace=True)
explode = (0, 0, 1.0)
plt.subplot(221)
zara_sentiment.transpose().plot(kind='barh',figsize=(18, 6))
plt.title('Sentiment Analysis 1', bbox={'facecolor':'0.8', 'pad':0})
plt.subplot(222)
zara_sentiment.plot(kind='pie',figsize=(18, 6),autopct='%1.1f%%',shadow=True,explode=explode)
plt.legend(bbox_to_anchor=(1, 1), loc=3, borderaxespad=0.)
plt.title('Sentiment Analysis 2', bbox={'facecolor':'0.8', 'pad':0})
plt.show()
```

```python
from sklearn import metrics

def timeseries_evaluacion_metrica_func(y_true, y_pred):
    
    def mean_absolute_percentage_error(y_true, y_pred): 
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    print('Resultados de la métrica de evaluación:-')
    print(f'MSE is : {metrics.mean_squared_error(y_true, y_pred)}')
    print(f'MSE is : {metrics.mean_absolute_error(y_true, y_pred)}')
    print(f'RMSE is : {np.sqrt(metrics.mean_squared_error(y_true, y_pred))}')
    print(f'MAPE is : {mean_absolute_percentage_error(y_true, y_pred)}')
    print(f'R2 is : {metrics.r2_score(y_true, y_pred)}',end='\n\n')
```

```python
timeseries_evaluacion_metrica_func(y_test, pred_xgb)
```

```python
timeseries_evaluacion_metrica_func(y_test, pred_rf)
```
