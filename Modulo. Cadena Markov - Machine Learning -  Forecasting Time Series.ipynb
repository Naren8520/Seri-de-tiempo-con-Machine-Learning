## Conceptos Básicos de las Cadenas de Markov: Definiciones Matemáticas

Las cadenas de Markov son modelos matemáticos que describen la evolución de sistemas que cambian aleatoriamente con el tiempo. Estos sistemas se caracterizan por un conjunto de estados y transiciones probabilísticas entre ellos.

**1. Definición de Cadenas de Markov:**

Una cadena de Markov es un proceso estocástico ${X_n}$ con un conjunto finito o contable de estados $S = {s_1, s_2, ..., s_n}$ y una matriz de transición $P = [P_{ij}]$, donde:

* **$X_n$:** Representa el estado del sistema en el tiempo $n$.
* **$S$:**  Es el conjunto de todos los estados posibles del sistema.
* **$P_{ij}$:** Es la probabilidad de transición del estado $s_i$ al estado $s_j$ en un solo paso de tiempo.

**2. Estados:**

Los estados son los diferentes valores que puede tomar la variable aleatoria $X_n$. Cada estado representa una condición o configuración específica del sistema.

**3. Transiciones:**

Las transiciones son los cambios de estado que ocurren en el sistema. Cada transición tiene una probabilidad asociada que determina la probabilidad de que ocurra.

**4. Probabilidad de Transición:**

La probabilidad de transición $P_{ij}$ representa la probabilidad de que el sistema pase del estado $s_i$ al estado $s_j$ en un solo paso de tiempo. Se define como:

$$P_{ij} = P(X_{n+1} = s_j | X_n = s_i),$$

donde:

* **$P(X_{n+1} = s_j | X_n = s_i)$:** Es la probabilidad condicional de que el sistema esté en el estado $s_j$ en el tiempo $n+1$ dado que estaba en el estado $s_i$ en el tiempo $n$.

**Propiedades de la Matriz de Transición:**

* **No negatividad:** $P_{ij} ≥ 0$ para todos $i, j$.
* **Suma de filas igual a 1:** $Σ_{j} P_{ij} = 1$ para todos $i$.

**Ejemplo:**

Consideremos un sistema que modela el clima de una ciudad con dos estados: "Soleado" (S) y "Lluvioso" (L). La matriz de transición es:

```
     S    L
S  [0.7  0.3]
L  [0.4  0.6] 
```

Esto significa que:

* La probabilidad de que el clima pase de soleado a soleado es 0.7.
* La probabilidad de que el clima pase de soleado a lluvioso es 0.3.
* La probabilidad de que el clima pase de lluvioso a soleado es 0.4.
* La probabilidad de que el clima pase de lluvioso a lluvioso es 0.6.

**En general:**

Las cadenas de Markov se definen por un conjunto de estados, transiciones y probabilidades de transición. La matriz de transición $P_{ij}$ resume las probabilidades de transición entre los estados. Estas definiciones matemáticas son esenciales para comprender y analizar el comportamiento de las cadenas de Markov.

## Tipos de Cadenas de Markov: Discreto vs. Continuo

Las cadenas de Markov se clasifican en dos tipos principales, dependiendo de cómo se maneja el tiempo:

**1. Cadenas de Markov de Tiempo Discreto:**

* **Tiempo discreto:** El tiempo se considera en pasos discretos, como por ejemplo, horas, días o semanas.
* **Transiciones en pasos específicos:** El sistema solo puede cambiar de estado en momentos específicos, definidos por los pasos de tiempo.
* **Ejemplos:**
    * El clima de una ciudad: Se observa el clima en momentos específicos (por ejemplo, cada día).
    * El número de clientes en una tienda: Se registra el número de clientes a intervalos regulares (por ejemplo, cada hora).
    * El precio de una acción: Se observa el precio de la acción al cierre de cada día.

**2. Cadenas de Markov de Tiempo Continuo:**

* **Tiempo continuo:** El tiempo se considera como un flujo continuo, sin pasos discretos.
* **Transiciones en cualquier momento:** El sistema puede cambiar de estado en cualquier momento.
* **Ejemplos:**
    * El número de clientes en un centro de llamadas: El número de clientes en espera puede cambiar en cualquier momento.
    * La temperatura de un horno: La temperatura puede variar continuamente a lo largo del tiempo.
    * La posición de un objeto en movimiento: La posición del objeto puede cambiar continuamente a medida que se mueve.

**Tabla comparativa:**

| Característica | Cadenas de Markov de Tiempo Discreto | Cadenas de Markov de Tiempo Continuo |
|---|---|---|
| Tiempo | Discreto (pasos) | Continuo (flujo) |
| Transiciones | En pasos específicos | En cualquier momento |
| Ejemplos | Clima, precio de acciones, número de clientes en una tienda | Número de clientes en un centro de llamadas, temperatura de un horno, posición de un objeto en movimiento |

**En general:**

* Las cadenas de Markov de tiempo discreto modelan sistemas que cambian en momentos específicos.
* Las cadenas de Markov de tiempo continuo modelan sistemas que cambian en cualquier momento.

La elección del tipo de cadena de Markov depende de la naturaleza del sistema que se está modelando. Si el sistema solo cambia en momentos específicos, se utiliza una cadena de Markov de tiempo discreto. Si el sistema puede cambiar en cualquier momento, se utiliza una cadena de Markov de tiempo continuo.

## Propiedades de las Cadenas de Markov: Definiciones Matemáticas

Las propiedades de las cadenas de Markov, como la irreductibilidad, la aperiódicidad y la recurrencia, son fundamentales para comprender su comportamiento a largo plazo. Estas propiedades se pueden definir matemáticamente de la siguiente manera:

**1. Irreductibilidad:**

* **Definición formal:** Una cadena de Markov con matriz de transición $P$ es irreductible si para cualquier par de estados $i$ y $j$, existe un entero $n \geq 1$ tal que $P_{ij}^n > 0$. 

* **Explicación:** Esto significa que desde cualquier estado $i$, es posible llegar al estado $j$ en un número finito de pasos, es decir, existe una secuencia de transiciones con probabilidad positiva que lleva de $i$ a $j$.

**2. Aperiodicidad:**

* **Definición formal:** Un estado $i$ de una cadena de Markov es aperiódico si el máximo común divisor (MCD) del conjunto de todos los enteros $n \geq 1$ para los cuales $P_{ii}^n > 0$ es 1.

* **Explicación:**  Esto significa que no hay un período común para las visitas al estado $i$. El sistema no se queda atrapado en un ciclo repetitivo. Una cadena de Markov es aperiódica si todos sus estados son aperiódicos.

**3. Recurrencia:**

* **Definición formal:** Un estado $i$ de una cadena de Markov es recurrente si la probabilidad de volver a ese estado en un número finito de pasos es 1, es decir, 
    
    $$P(T_i < \infty) = 1,$$

    donde $T_i$ es el tiempo de primer retorno al estado $i$.

* **Explicación:**  Esto significa que el sistema eventualmente volverá a visitar el estado $i$. Un estado es transitorio si no es recurrente.

**En resumen:**

* **Irreductibilidad:**  Todos los estados son accesibles desde cualquier otro estado.
* **Aperiodicidad:**  No hay un período común para las visitas a los estados.
* **Recurrencia:**  El sistema eventualmente volverá a visitar los estados recurrentes.

**Nota:** 

Estas definiciones matemáticas proporcionan una base sólida para comprender las propiedades de las cadenas de Markov. Son esenciales para analizar el comportamiento a largo plazo del sistema y para realizar predicciones precisas sobre su evolución.

## La Matriz de Transición: El Corazón de las Cadenas de Markov

La matriz de transición es el elemento fundamental de una cadena de Markov. Describe las probabilidades de transición entre los estados del sistema. Para construir la matriz de transición, necesitamos identificar los estados posibles del sistema y las probabilidades de pasar de un estado a otro en un solo paso de tiempo.

**Definición Matemática:**

La matriz de transición $P$ es una matriz cuadrada donde cada elemento $P_{ij}$ representa la probabilidad de transición del estado $i$ al estado $j$ en un solo paso de tiempo. La matriz $P$ tiene las siguientes propiedades:

* **No negatividad:** $P_{ij} ≥ 0$ para todos $i, j$.
* **Suma de filas igual a 1:** $Σ_{j} P_{ij} = 1$ para todos $i$.

**Construcción de la Matriz de Transición:**

1. **Identificar los estados:**  Definir el conjunto de todos los estados posibles del sistema.
2. **Determinar las probabilidades de transición:**  Para cada par de estados $(i, j)$, calcular la probabilidad $P_{ij}$ de pasar del estado $i$ al estado $j$ en un solo paso de tiempo.
3. **Organizar la matriz:**  Organizar las probabilidades de transición en una matriz cuadrada, donde las filas representan los estados de origen y las columnas representan los estados de destino.

## Calculando Probabilidades de Transición en Múltiples Pasos: Un Viaje a través del Tiempo en Cadenas de Markov

En las cadenas de Markov, a menudo necesitamos calcular la probabilidad de pasar de un estado a otro en un número específico de pasos. Esto es crucial para comprender la evolución del sistema a lo largo del tiempo. Para lograr esto, podemos utilizar la matriz de transición y la potencia de la matriz.

**Definición Matemática:**

La probabilidad de pasar del estado $i$ al estado $j$ en $n$ pasos se denota como $P_{ij}^n$ y se calcula como la entrada $(i, j)$ de la matriz de transición elevada a la potencia $n$:

$$P_{ij}^n = (P^n)_{ij}$$

donde $P^n$ es la matriz de transición elevada a la potencia $n$.

**Cálculo en Python:**

Para calcular las probabilidades de transición en múltiples pasos en Python, podemos utilizar la función `numpy.linalg.matrix_power()`. Esta función calcula la potencia de una matriz.

**Ejemplo:**

Consideremos el ejemplo del clima de una ciudad con estados "Soleado" (S) y "Lluvioso" (L) y la siguiente matriz de transición:


```
     S    L
S  [0.7  0.3]
L  [0.4  0.6] 
```

````python
import numpy as np

# Matriz de transición
P = np.array([[0.7, 0.3],
              [0.4, 0.6]])

# Calcular la probabilidad de pasar de Soleado a Lluvioso en 3 pasos
P_3 = np.linalg.matrix_power(P, 3)
probabilidad_SL_3 = P_3[0, 1]

print(f"Probabilidad de pasar de Soleado a Lluvioso en 3 pasos: {probabilidad_SL_3}")
````

**Explicación:**

* `np.linalg.matrix_power(P, 3)` calcula la matriz de transición elevada a la potencia 3, es decir, P^3.
* `P_3[0, 1]` accede a la entrada (0, 1) de la matriz P^3, que representa la probabilidad de pasar del estado 0 (Soleado) al estado 1 (Lluvioso) en 3 pasos.

**Otro Ejemplo:**

````python
# Calcular la probabilidad de pasar de Lluvioso a Soleado en 2 pasos
P_2 = np.linalg.matrix_power(P, 2)
probabilidad_LS_2 = P_2[1, 0]

print(f"Probabilidad de pasar de Lluvioso a Soleado en 2 pasos: {probabilidad_LS_2}")
````

Para calcular las probabilidades de transición en múltiples pasos, podemos utilizar la potencia de la matriz de transición. En Python, podemos usar la función `numpy.linalg.matrix_power()` para realizar este cálculo de forma eficiente. Esto nos permite analizar la evolución del sistema a lo largo del tiempo y comprender su comportamiento a largo plazo.

## Estados Estacionarios: El Equilibrio en las Cadenas de Markov

Los estados estacionarios en una cadena de Markov representan un estado de equilibrio donde la distribución de probabilidad de los estados permanece constante a lo largo del tiempo. En otras palabras, si el sistema alcanza un estado estacionario, la probabilidad de estar en cada estado no cambia con el paso del tiempo.

**Definición Matemática:**

Un vector de probabilidad $π = (π_1, π_2, ..., π_n)$ es un estado estacionario de una cadena de Markov con matriz de transición $P$ si satisface la siguiente ecuación:

$$π = πP$$

Esto significa que la distribución de probabilidad $π$ permanece sin cambios después de una transición.

**Cómo Identificar Estados Estacionarios:**

1. **Resolver el sistema de ecuaciones:** La ecuación $π = πP$ es un sistema de ecuaciones lineales. Para encontrar los estados estacionarios, debemos resolver este sistema.
2. **Condición de normalización:**  La suma de las componentes del vector de probabilidad $π$ debe ser igual a 1:

    $$Σ_{i} π_i = 1$$

3. **Solución única:**  En general, una cadena de Markov puede tener uno o más estados estacionarios. Si la cadena es irreductible y aperiódica, tendrá un único estado estacionario.



**Ejemplo:**

Consideremos el ejemplo del clima de una ciudad con la siguiente matriz de transición:

```
     S    L
S  [0.7  0.3]
L  [0.4  0.6]
```

Para encontrar el estado estacionario, resolvemos la ecuación π = πP:

```
π_S = 0.7π_S + 0.4π_L
π_L = 0.3π_S + 0.6π_L
```

Además, tenemos la condición de normalización:

```
π_S + π_L = 1
```

Resolviendo este sistema de ecuaciones, obtenemos:

```
π_S = 4/7
π_L = 3/7
```

Por lo tanto, el estado estacionario de esta cadena de Markov es (4/7, 3/7), lo que significa que a largo plazo, el clima de la ciudad tendrá un 4/7 de probabilidad de ser soleado y un 3/7 de probabilidad de ser lluvioso.

**En general:**

Los estados estacionarios representan un estado de equilibrio en las cadenas de Markov. Para identificarlos, debemos resolver un sistema de ecuaciones lineales que relaciona la distribución de probabilidad con la matriz de transición. Los estados estacionarios son importantes para comprender el comportamiento a largo plazo del sistema y para realizar predicciones sobre su evolución.

## Encontrando Estados Estacionarios

Vamos a encontrar el estado estacionario de una cadena de Markov que modela el comportamiento de un usuario en un sitio web.

**Ejemplo:**

Supongamos que un servidor web puede estar en uno de tres estados:

* **Disponible (A):** El servidor está funcionando correctamente y puede atender solicitudes.
* **Ocupado (O):** El servidor está procesando una solicitud.
* **Falla (F):** El servidor está fuera de servicio.

La matriz de transición para este sistema es:

```
    A    O    F
A  [0.8  0.1  0.1]
O  [0.6  0.3  0.1]
F  [0.2  0.0  0.8]
```




````python
import numpy as np
import numpy.linalg as LA

# Matriz de transición
P = np.array([[0.8, 0.1, 0.1],
              [0.6, 0.3, 0.1],
              [0.2, 0.0, 0.8]])

print(P)
````

````python
# Encontrar el estado estacionario
eigenvalues, eigenvectors = LA.eig(P.T)  # Transponer P para obtener el estado estacionario correcto
stationary_state = eigenvectors[:, np.argmax(eigenvalues)]  # Obtener el vector propio asociado al valor propio 1
stationary_state = stationary_state / np.sum(stationary_state)  # Normalizar el vector

# Imprimir el estado estacionario
print("Estado estacionario:")
print(stationary_state)
````

Explicación:

1. Definir la matriz de transición: Se define la matriz P que representa las probabilidades de transición entre los estados.

2. Encontrar el estado estacionario:

* `LA.eig(P.T)` calcula los valores propios y vectores propios de la matriz transpuesta de P.

* `np.argmax(eigenvalues)` encuentra el índice del valor propio más cercano a 1 (el valor propio dominante).

* `eigenvectors[:, np.argmax(eigenvalues)]` selecciona el vector propio asociado al valor propio dominante.

* `stationary_state / np.sum(stationary_state)` normaliza el vector para que la suma de sus componentes sea 1.

3. Imprimir el estado estacionario: Se imprime el vector `stationary_state` que representa el estado estacionario de la cadena de Markov.

Interpretación:

El estado estacionario encontrado indica que a largo plazo, el servidor web tendrá aproximadamente un 58.3% de probabilidad de estar disponible, un 8.3% de probabilidad de estar ocupado y un 33.33% de probabilidad de estar en falla.



# Planificación de la Demanda con Cadenas de Markov

### Paso 1: Preparar los Datos

Supongamos que tienes datos históricos de ventas mensuales de una empresa, organizados en una lista:

````python
ventas = [100, 120, 110, 130, 140, 150, 130, 120, 110, 100, 120, 130]
````

### Paso 2: Discretizar los Datos

Para aplicar la cadena de Markov, necesitamos discretizar los datos de ventas en un número finito de estados. Podemos hacerlo creando rangos de ventas y asignando cada valor de venta a un estado específico.

````python
# Definir los estados de demanda
estados = ["Baja", "Media", "Alta"]

# Discretizar las ventas
ventas_discretizadas = []
for venta in ventas:
    if venta <= 110:
        ventas_discretizadas.append("Baja")
    elif venta <= 130:
        ventas_discretizadas.append("Media")
    else:
        ventas_discretizadas.append("Alta")
````

````python
ventas_discretizadas
````

### Paso 3: Construir la Matriz de Transición

La matriz de transición representa las probabilidades de pasar de un estado de demanda a otro. Podemos construirla contando las transiciones entre estados en los datos discretizados.

````python
import numpy as np

# Crear una matriz de transición inicializada con ceros
P = np.zeros((len(estados), len(estados)))

# Contar las transiciones
for i in range(len(ventas_discretizadas) - 1):
    estado_actual = ventas_discretizadas[i]
    estado_siguiente = ventas_discretizadas[i + 1]
    indice_actual = estados.index(estado_actual)
    indice_siguiente = estados.index(estado_siguiente)
    P[indice_actual, indice_siguiente] += 1

# Normalizar la matriz para obtener probabilidades
for i in range(len(estados)):
    P[i] = P[i] / np.sum(P[i])
````

````python
P
````

### Paso 4: Calcular el Estado Estacionario

El estado estacionario representa la distribución de probabilidad de los estados de demanda a largo plazo. Podemos calcularlo usando la función `numpy.linalg.eig` para encontrar el vector propio asociado al valor propio 1.

````python
# Encontrar el estado estacionario
eigenvalues, eigenvectors = np.linalg.eig(P.T)  # Transponer P para obtener el estado estacionario correcto
stationary_state = eigenvectors[:, np.argmax(eigenvalues)]  # Obtener el vector propio asociado al valor propio 1
stationary_state = stationary_state / np.sum(stationary_state)  # Normalizar el vector

# #estados = ["Baja", "Media", "Alta"]
# Imprimir el estado estacionario
print("Estado estacionario:")
print(stationary_state)
````

### Paso 5: Planificar la Demanda
El estado estacionario nos proporciona una estimación de la probabilidad de cada estado de demanda a largo plazo. Podemos usar esta información para planificar la producción, el inventario y otras operaciones de la empresa.

````python
# Ejemplo: Planificación de la producción
produccion_planificada = [
    int(stationary_state[0] * 100),  # Producción para demanda baja
    int(stationary_state[1] * 120),  # Producción para demanda media
    int(stationary_state[2] * 140),  # Producción para demanda alta
]

print("Producción planificada:")
print(produccion_planificada)
````

1. Preprocesar los datos
2. Discretizar los precios en estados
3. Calcular la matriz de transición
4. Realizar la predicción

````python
import pandas as pd
import numpy as np
from datetime import timedelta
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split



# Discretizar los precios en estados con etiquetas descriptivas
n_states = 5
labels = ["Muy Bajo", "Bajo", "Medio", "Alto", "Muy Alto"]
data['State'] = pd.qcut(data['Price'], q=n_states, labels=labels)

# Crear características adicionales
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year
data['PriceDiff'] = data['Price'].diff()
data['PriceRollingMean'] = data['Price'].rolling(window=3).mean()
data['PriceRollingStd'] = data['Price'].rolling(window=3).std()

# Eliminar filas con NaN
data = data.dropna()

# Preparar los datos para el modelo Random Forest
X = data[['Price', 'Month', 'Year', 'PriceDiff', 'PriceRollingMean', 'PriceRollingStd']]
y = data['State']

# Escalar las características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Entrenar el modelo Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluar el modelo
train_accuracy = model.score(X_train, y_train)
test_accuracy = model.score(X_test, y_test)
print(f"Precisión en entrenamiento: {train_accuracy:.4f}")
print(f"Precisión en prueba: {test_accuracy:.4f}")

# Función para predecir el próximo estado usando el modelo Random Forest
def predict_next_state_rf(current_features):
    scaled_features = scaler.transform([current_features])
    probabilities = model.predict_proba(scaled_features)[0]
    return labels[np.random.choice(len(labels), p=probabilities)]

# Función para convertir estado a precio
def state_to_price(state):
    price_ranges = pd.qcut(data['Price'], q=n_states)
    state_index = labels.index(state)
    price_range = price_ranges.cat.categories[state_index]
    return price_range.mid

# Realizar predicciones
last_row = data.iloc[-1]
predictions = []

for i in range(5):
    next_date = last_row['Date'] + timedelta(days=31)
    current_features = [
        last_row['Price'],
        next_date.month,
        next_date.year,
        last_row['PriceDiff'],
        last_row['PriceRollingMean'],
        last_row['PriceRollingStd']
    ]
    next_state = predict_next_state_rf(current_features)
    next_price = state_to_price(next_state)
    predictions.append((next_date, next_state, next_price))
    
    # Actualizar last_row para la siguiente iteración
    last_row = pd.Series({
        'Date': next_date,
        'Price': next_price,
        'Month': next_date.month,
        'Year': next_date.year,
        'PriceDiff': next_price - last_row['Price'],
        'PriceRollingMean': (last_row['PriceRollingMean'] * 2 + next_price) / 3,
        'PriceRollingStd': np.std([last_row['Price'], last_row['PriceRollingMean'], next_price])
    })

# Imprimir predicciones
print("\nPredicciones para los próximos 5 meses:")
for date, state, price in predictions:
    print(f"Fecha: {date.strftime('%Y-%m-%d')}, Estado: {state}, Precio pronosticado: {price:.2f}")

# Graficar la serie temporal original y las predicciones
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Price'], label='Datos históricos')
pred_dates = [p[0] for p in predictions]
pred_prices = [p[2] for p in predictions]
plt.plot(pred_dates, pred_prices, 'ro-', label='Predicciones')
plt.title('Serie Temporal y Predicciones (Random Forest + Markov)')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Graficar la importancia de las características
feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feature_importance)
plt.title('Importancia de las Características en el Random Forest')
plt.xlabel('Importancia')
plt.ylabel('Característica')
plt.tight_layout()
plt.show()
````

````python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import graphviz
````

````python
# Matriz de transición de ejemplo
transition_matrix = np.array([[0.7, 0.3, 0.0],
                              [0.2, 0.5, 0.3],
                              [0.1, 0.2, 0.7]])

G = nx.DiGraph()

# Agregar nodos
for i in range(transition_matrix.shape[0]):
    G.add_node(str(i))

# Agregar aristas con pesos
for i in range(transition_matrix.shape[0]):
    for j in range(transition_matrix.shape[1]):
        if transition_matrix[i, j] > 0:
            G.add_edge(str(i), str(j), weight=transition_matrix[i, j])


# Crear un objeto Dot
dot = graphviz.Graph(comment='Matriz de Transición')

# Agregar nodos y aristas al objeto Dot
for u, v, d in G.edges(data=True):
    dot.edge(u, v, label=str(round(d['weight'], 2)))

# Guardar el gráfico como un archivo PNG
dot.render('transition_matrix.gv', view=True)
````
